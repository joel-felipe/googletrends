```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["digital Marketing"])
data = trends.interest_by_region()
data = data.sort_values(by="digital Marketing", ascending=False)
data = data.head(15)
print(data)
```

```python
data.reset_index().plot(x="geoName", y="digital Marketing",
                       figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['digital Marketing'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data['digital Marketing'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Digital Marketing', fontweight='bold')
plt.xlabel('Year')
plt.show()
```
